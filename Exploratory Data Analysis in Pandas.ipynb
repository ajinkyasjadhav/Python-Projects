# Objective:
Are the certain continents are grown faster than others & in which ways...?

***Grouping all the Continents***

```python
df.groupby('Continent').mean().sort_values( by = '2022 Population', ascending = False)
```

***To plot the graph for considering year wise popluation columns mention the columns as below***

```python
df2 = df.groupby('Continent')['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population'].mean().sort_values( by = '2022 Population', ascending = False)
df2
```

```python
df.columns
```

Tranpose (Interchange the headers in the Dataframe)

```python
df3 = df2.transpose()
df3
```

This is how the continents have grown

```python
df3.plot()
```
